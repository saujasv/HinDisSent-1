```python
import xml.etree.ElementTree as et 

xtree = et.parse("ReviewSentimentDataset-AspectTermExtraction.xml")
xroot = xtree.getroot()
```

```python
sentences = []

for node in xroot:
    sentence = {}
    sentence_attribs = []
    for child in node:
        if child.tag == "aspectTerms":
            for baby in child:
                print(baby.attrib)
                sentence_attribs.append(baby.attrib)
        elif child.tag == "text":
            print("Text:",child.text)
            sentence['text'] = child.text
    sentence['aspectTerms'] = sentence_attribs
    sentences.append(sentence)
    
```

# Data
Dataset taken from the Aspect Based Sentiment Reviews 

## Structure

- `sentences` contains all the sentences. 
- For each `sentence` in `sentences`
    - `sentence['text']` contains the text or the raw sentence itself
    - `sentence['aspectTerms']` is a list of all the aspect Terms in the sentence
    - For each `aspectTerm` in `sentence['aspectTerms']` contains 
        - `aspectTerm['polarity']`: Polarity : pos/neg/neu
        - `aspectTerm['term']`: The term attribute
        - `aspectTerm['from']`: Starting character of the term block 
        - `aspectTerm['to']`: Ending character of the aspectTerm.
